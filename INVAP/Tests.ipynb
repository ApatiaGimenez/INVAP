### Leer csv


```python
import pandas as pd

# Define the file path
csv_file_path = 'CSVEnc.csv'

# Read the CSV file into a DataFrame with specific options
df = pd.read_csv(csv_file_path, delimiter=',', header=0)

# Display the DataFrame
print(df)
```

### Pruebas estadisticas según genero

```python
from scipy import stats
from scipy.stats import ttest_ind, t
```

```python
# Filtrar los datos para separar las mujeres y el resto
women_data = df[df['Identidad de género'] == 'Mujer']
other_data = df[df['Identidad de género'] != 'Mujer']
```

```python
# Obtener el índice de la columna 11 (indexación basada en cero)
start_column_index = 11

# Definir el nivel de confianza
confidence_level = 0.95

# Iterar sobre los índices de las columnas seleccionadas
for column_index in range(start_column_index, len(df.columns)):
    # Realizar la prueba de diferencia de medias (prueba t de Student)
    t_stat, p_value = ttest_ind(women_data.iloc[:, column_index], other_data.iloc[:, column_index], equal_var=False)

    # Calcular las medias de ambos grupos
    mean_women = women_data.iloc[:, column_index].mean()
    mean_other = other_data.iloc[:, column_index].mean()

    # Calcular los intervalos de confianza
    mean_diff = mean_women - mean_other
    std_err_diff = ((women_data.iloc[:, column_index].std() ** 2) / len(women_data) + (other_data.iloc[:, column_index].std() ** 2) / len(other_data)) ** 0.5
    margin_of_error = t.ppf((1 + confidence_level) / 2, len(women_data) + len(other_data) - 2) * std_err_diff

    # Calcular los límites del intervalo de confianza
    lower_ci = mean_diff - margin_of_error
    upper_ci = mean_diff + margin_of_error

    # Obtener el nombre de la columna
    column_name = df.columns[column_index]

    # Imprimir los resultados
    print(f"Prueba t para la columna en el índice {column_index}:")
    print("  Nombre de la columna:", column_name)
    print("  Media de mujeres:", mean_women)
    print("  Media del resto:", mean_other)
    print("  Estadístico t:", t_stat)
    print("  Valor p:", p_value)
    print("  Intervalo de confianza para la diferencia de medias:", (lower_ci, upper_ci))
```

### Pruebas estadísticas según educación

```python
# Filtrar los datos para separar por nivel educativo
education_levels_of_interest = ["Primario completo", "Secundario incompleto", "Secundario completo", "Terciario incompleto", "Terciario completo", "Universitario incompleto"]
education_data = df[df['¿Cuál es el máximo nivel educativo que alcanzaste?'].isin(education_levels_of_interest)]
other_education_data = df[~df['¿Cuál es el máximo nivel educativo que alcanzaste?'].isin(education_levels_of_interest)]
```

```python
# Obtener el índice de la columna 11 (indexación basada en cero)
start_column_index = 11

# Definir el nivel de confianza
confidence_level = 0.95

# Iterar sobre los índices de las columnas seleccionadas
for column_index in range(start_column_index, len(df.columns)):
    # Realizar la prueba de diferencia de medias (prueba t de Student)
    t_stat, p_value = ttest_ind(education_data.iloc[:, column_index], other_education_data.iloc[:, column_index], equal_var=False)

    # Calcular las medias de ambos grupos
    mean_education = education_data.iloc[:, column_index].mean()
    mean_other_education = other_education_data.iloc[:, column_index].mean()

    # Calcular los intervalos de confianza
    mean_diff = mean_education - mean_other_education
    std_err_diff = ((education_data.iloc[:, column_index].std() ** 2) / len(education_data) + (other_education_data.iloc[:, column_index].std() ** 2) / len(other_education_data)) ** 0.5
    margin_of_error = t.ppf((1 + confidence_level) / 2, len(education_data) + len(other_education_data) - 2) * std_err_diff

    # Calcular los límites del intervalo de confianza
    lower_ci = mean_diff - margin_of_error
    upper_ci = mean_diff + margin_of_error

    # Obtener el nombre de la columna
    column_name = df.columns[column_index]

    # Imprimir los resultados
    print(f"Prueba t para la columna en el índice {column_index}:")
    print("  Nombre de la columna:", column_name)
    print("  Media de personas con nivel educativo de interés:", mean_education)
    print("  Media del resto de personas:", mean_other_education)
    print("  Estadístico t:", t_stat)
    print("  Valor p:", p_value)
    print("  Intervalo de confianza para la diferencia de medias:", (lower_ci, upper_ci))
```
